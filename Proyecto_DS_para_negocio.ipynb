**Implementación**

El siguiente código define una función llamada **graficos_eda_categoricos** que crea gráficos de barras horizontales para variables categóricas en un conjunto de datos. Esta función es útil para realizar un análisis exploratorio de datos (EDA) rápido en variables categóricas, creando visualizaciones de la distribución de cada variable en un solo gráfico.

* **def graficos_eda_categoricos(cat):** Define la función que toma un parámetro cat, que se espera que sea un DataFrame de pandas con variables categóricas.
* **from math import ceil**: Importa la función ceil del módulo math. Esta función se usa para redondear hacia arriba.
* **filas = ceil(cat.shape[1] / 2)**: Calcula el número de filas necesarias para los subplots. Divide el número de columnas en cat por 2 y redondea hacia arriba.
* **f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))**: Crea una figura con subplots. El número de filas es filas, y siempre hay 2 columnas. El tamaño de la figura es de 16 unidades de ancho y 6 unidades de alto por cada fila.
* **ax = ax.flat**: Convierte el array 2D de ejes en un array 1D para facilitar la iteración.

* El bucle **for** itera sobre cada variable en el DataFrame **cat**:

 1. **cat[variable].value_counts().plot.barh(ax = ax[cada])**: Crea un gráfico de barras horizontal para cada variable, mostrando el conteo de cada categoría.
 2. **ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")**: Establece el título de cada subplot como el nombre de la variable, en negrita y tamaño 12.
 3. **ax[cada].tick_params(labelsize = 12)**: Establece el tamaño de las etiquetas de los ejes en 12.

```python
def graficos_eda_categoricos(cat):

    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)
```

Ahora ejecutamos la funcion definida

### EDA Variables Numéricas

En esta etapa analizaremos las variables numericas a traves de valores estadisticos, como media, mediana, moda, etc.

La función que se creará a continuación toma un parámetro **num**, que se espera que sea un DataFrame de pandas con variables numéricas.

* **num.describe().T:**
 * **describe()** calcula estadísticos descriptivos básicos para cada columna numérica.
 * **.T** transpone el resultado, convirtiendo filas en columnas y viceversa.

* **estadisticos['median'] = num.median()**:
Calcula la mediana para cada columna y la añade como una nueva
 columna llamada 'median' al DataFrame de estadísticos.

* **estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]**:
 * Reordena las columnas del DataFrame.
 * El orden resultante es: count, mean, median, std, min, 25%, 50%, 75%, max.
 * Esto coloca la mediana (índice 8) justo después de la media (índice 1).

* **return(estadisticos)**:
Devuelve el DataFrame de estadísticos modificado.

```python
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)
```

Ejecutamos la función:

**.select_dtypes('number')** selecciona todas las columnas del DataFrame que contienen datos numéricos. Esto incluye tipos como int64, float64, etc.

```python
estadisticos_cont(df.select_dtypes('number'))
```

En esencia, esta línea está calculando el salario anual multiplicando el salario mensual por 12.

Explicación:

* Se definen cuatro condiciones basadas en rangos salariales anuales:
1. Salario <= 30,000
2. 30,000 < Salario <= 50,000
3. 50,000 < Salario <= 75,000
4. Salario > 75,000

* Para cada rango salarial, se define un cálculo correspondiente:
1. 16.1% del salario para el primer rango
2. 19.7% del salario para el segundo rango
3. 20.4% del salario para el tercer rango
4. 21% del salario para el cuarto rango

* Se utiliza **np.select()** para aplicar estos cálculos condicionalmente, creando una nueva columna llamada 'impacto_abandono'.

* El valor predeterminado de -999 probablemente se usa para marcar datos que no caen en ninguna de las categorías definidas, lo que podría indicar errores en los datos o casos especiales que requieren atención adicional.

Con esto se calcula el impacto financiero del abandono de empleados basado en sus salarios. Los porcentajes crecientes sugieren que el costo de reemplazo o el impacto del abandono aumenta proporcionalmente con el salario.

```python
# Calculamos el impacto económico de cada empleado si deja la empresa

#Lista de condiciones
condiciones = [(df['salario_ano'] <= 30000),
               (df['salario_ano'] > 30000) & (df['salario_ano'] <= 50000),
               (df['salario_ano'] > 50000) & (df['salario_ano'] <= 75000),
               (df['salario_ano'] > 75000)]

#Lista de resultados
resultados = [df.salario_ano * 0.161, df.salario_ano * 0.197, df.salario_ano * 0.204, df.salario_ano * 0.21]

#Aplicamos select
df['impacto_abandono'] = np.select(condiciones,resultados, default = -999)

# Mostramos el data frame
df
```

Linea 3:  convertir variables categóricas en una forma que pueda ser proporcionada a algoritmos de aprendizaje automático para hacer mejores predicciones.

- Creamos una instancia de la clase OneHotEncoder de la biblioteca scikit-learn.
- OneHotEncoder(): Esta es la clase que se está instanciando.
- sparse = False: Este parámetro indica que queremos que el resultado de la codificación sea una matriz densa (array) en lugar de una matriz dispersa (sparse matrix).
- Una matriz densa es más fácil de trabajar con pandas, aunque puede usar más memoria.

El codificador one-hot transforma cada categoría de una variable en una nueva columna binaria (0 o 1). Por ejemplo, si tienes una variable "Color" con categorías "Rojo", "Azul" y "Verde", el codificador one-hot la transformaría en tres columnas: "Color_Rojo", "Color_Azul" y "Color_Verde", donde cada fila tendría un 1 en la columna correspondiente y 0 en las demás.

Linea 4: identifcamos las variables categoricas.
- fit() analiza los datos en cat para determinar todas las categorías únicas en cada columna.
- No transforma los datos, solo "aprende" las categorías.
- Después de esto, ohe conoce todas las posibles categorías para cada variable.

Linea 5: aplicamos transformaciones.
- transform() aplica la codificación one-hot a los datos en cat.
- Crea una nueva columna binaria (0 o 1) para cada categoría de cada variable original.
- cat_ohe es una matriz numpy, no un DataFrame de pandas.

Linea 6: Renombramos las variables.
- pd.DataFrame() convierte la matriz numpy cat_ohe en un DataFrame de pandas.
- ohe.get_feature_names_out() genera nombres para las nuevas columnas basados en las columnas originales y sus categorías.
- input_features = cat.columns proporciona los nombres de las columnas originales para generar los nuevos nombres.
- reset_index(drop = True) asegura que el nuevo DataFrame tenga un índice numérico simple, descartando cualquier índice del DataFrame original.

```python
# 1. Importamos librería
from sklearn.preprocessing import OneHotEncoder

# 2. Categóricas
cat = df_ml.select_dtypes('O')

# 3. Instanciamos
ohe = OneHotEncoder(sparse_output=False)

# 4. Entrenamos
ohe.fit(cat)

# 5. Aplicamos
cat_ohe = ohe.transform(cat)

# 6. Ponemos los nombres
cat_ohe = pd.DataFrame(cat_ohe, columns = ohe.get_feature_names_out(input_features = cat.columns)).reset_index(drop = True)
```
